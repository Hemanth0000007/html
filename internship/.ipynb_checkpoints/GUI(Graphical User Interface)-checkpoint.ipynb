```python
def sort_words(sentence):
    # Split the sentence into words
    words = sentence.split()

    # Sort words based on length in ascending order
    words.sort(key=lambda word: (len(word), word))
    
    return words

# Example usage
sentence = input()
sorted_words = sort_words(sentence)
print(*sorted_words)
```

```python
def s_w(s):
    w = s.split()
    w.sort(key=lambda w: (len(w), w))
    return w
s= input()
so_w = s_w(s)
print(*so_w)
```
