# QUEUE DATA STRUCTURE
- Queue is a data structure that follows FIFO(first come first out) or FCFS(first come first serve) principle
- the elements that enters into the queue is the first one to go out
- in queue,the insertions will be made from end and deletions will be made from beginning 
- in a normal queue we define the following operations
 - enque() --> inserting a new element into queue
 - deque() --> deleting an element from the queue
 - get_front() --> viewing the element at front
 - get_rear() --> viewing the last element of the queue
 - get_size() --> getting the length
 - view_queue() --> viewing all elements in the queue
 - is_empty() --> to check if queue has no elements
- in python,queues can be implemented using
 - a list 
 - a collection deque
 - a linked list
 

# 1.1 Queue using lists

```python
q = []

def is_empty():
    if len(q) == 0:
        return True
    else:
        return False

def enque(val):
    global q
    q.append(val)
    
def deque():
    global q
    if is_empty():
        print('Queue is empty!')
    else:
        x = q.pop(0)
        print(f"{x} is deleted")

def get_front():
    global q
    if is_empty():
        return -1
    else:
        return q[0]
              
def get_rear():
    global q
    if is_empty():
        return -1
    else:
        return q[-1] 

def get_size():
    global q
    if is_empty():
        print('Queue is empty!')
    else:
        return len(q) 
              
def view_queue():
    global q
    if is_empty():
        print('Queue is empty!')
    else:
        for i in q:
            print(i,end = '\n')
        
while 1:
    ch = int(input("""ENTER
    1 . Enque
    2 . Deque
    3 . is_empty
    4 . get_front
    5 . get_rear
    6 . get_size
    7 . view_queue:"""))
    if ch == 1:
        val = int(input('Enter an element to insert: '))
        enque(val)
    elif ch == 2:
        deque()
    elif ch == 3:
        if is_empty():
            print('Queue is empty!')
        else :
              print("Queue is not empty")
    elif ch == 4:
        if is_empty():
            print('Queue is empty!')
        else :
              print(get_front())
    elif ch == 5:
        if is_empty():
            print('Queue is empty!')
        else :
              print(get_rear())
    elif ch == 6:
        if is_empty():
            print('Queue is empty!')
        else :
            print(get_size())
    elif ch == 7:
        if is_empty():
            print('Queue is empty!')
        else :
             print(view_queue())
    else:
        print("THANK YOU")
        break
              
        
```
